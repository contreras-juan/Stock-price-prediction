```python
data = pd.read_csv("/content/drive/MyDrive/IA/Proyecto/data.csv", sep=";", index_col="Date")
tickers = data.columns
```

```python
len(tickers)
```

```python
tickers = tickers[tickers!="ABMD"]
```

```python
import matplotlib.pyplot as plt

for k in tickers:
  df = pd.read_csv("/content/drive/MyDrive/IA/Proyecto/Export/Data/" + k + ".csv", sep=";", index_col="Date")


  df.plot(figsize=(10,6))
  plt.xticks(rotation=45)
  plt.xlabel(None)
  plt.ylabel('Stock Price',size=15)
  plt.title(k,size=15)
  plt.show()
```

```python
stocks = pd.DataFrame()

for k in tickers:
  df = pd.read_csv("/content/drive/MyDrive/IA/Proyecto/Export/Data/" + k + ".csv", sep=";", index_col="Date")

  df.columns = [k + "_actual", k + "_predicted"]

  stocks = pd.concat([stocks, df], axis=1)
```

```python
stocks.head()
```

```python
from statistics import mean

def mae(actual, predicted):
  
  return mean(abs(actual - predicted))
```

```python
MAE = pd.DataFrame.from_dict(dict([(ticker, mae(stocks[ticker + "_actual"], stocks[ticker + "_predicted"])) for ticker in tickers]), orient='index', columns=["mae"])
```

```python
def plot_best_k(n_best, data):

  sorted = list(data.dropna().sort_values("mae").index)

  tickers_to_plot = sorted[0:n_best]

  for k in tickers_to_plot:
      
      stocks[[k + "_actual", k + "_predicted"]].plot(figsize=(10,6))
      plt.xticks(rotation=45)
      plt.xlabel(None)
      plt.ylabel('Stock Price',size=15)
      plt.title(k,size=15)
      plt.show()
```

```python
MAE.sort_values("mae")
```

```python
plot_best_k(20, MAE)
```
