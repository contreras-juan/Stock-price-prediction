```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
tickers = os.listdir('../Export/Predictions/One-step/')
```

```python
for k in tickers:

    df = pd.read_csv('../Export/Predictions/One-step/'+k, sep=";", index_col="Date")

    df.index = [x[0:10] for x in df.index]
    df.index = pd.to_datetime(df.index)

    df.plot(figsize=(10,6))
    plt.xticks(rotation=45)
    plt.xlabel(None)
    plt.ylabel('Stock Price',size=15)
    plt.title(k,size=15)
    plt.show()
```
